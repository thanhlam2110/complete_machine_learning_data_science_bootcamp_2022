# Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]
# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #width, height
# 3. Plot data
ax.plot(x,y)
# 4. Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")
# 5. Save & show 
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays
We want:
- Line plot
- Scatter plot
- Bar plot
- Histogram
- Subplots

```python
import numpy as np
```

```python
# 1. Create some data
x = np.linspace(0,10,100)
x
```

```python
# 2. Plot the data (line plot)
fig, ax = plt.subplots(figsize=(5,5)) #width, height
ax.plot(x,x**2);
```

```python
# 3. Plot same data to make a scatter
fig, ax = plt.subplots(figsize=(5,5)) #width, height
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots(figsize=(5,5)) #width, height
ax.scatter(x, np.sin(x));
```

```python
# 4. Bar plot (Make plot from dictionary)
nut_butter_prices = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter":12}
fig, ax = plt.subplots(figsize=(5,5)) #width, height
ax.bar(nut_butter_prices.keys(),height=nut_butter_prices.values());
ax.set(title="Lam's Nut Butter Store",
      ylabel="Price($)");
```

```python
fig, ax = plt.subplots(figsize=(5,5)) #width, height
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax.set(title="Lam's Nut Butter Store",
      xlabel="Price($)");
```

```python
# 5. Histogram
x = np.random.randn(1000) #"standard normal" distribution
fig, ax = plt.subplots(figsize=(5,5)) #width, height
ax.hist(x);
```

```python
# 6. Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# 7. Subplot option 2
fig, ax = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```
